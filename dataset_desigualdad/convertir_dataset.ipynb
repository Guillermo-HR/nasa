```python
import geopandas as gpd
```

```python
mapa = gpd.read_file('M03_01_La_Paz.shp')

mapa.to_csv('M03_01_La_Paz.csv', index=False)
```

```python
import folium

# Crear un mapa centrado en La Paz
m = folium.Map(location=[24.16, -110.30], zoom_start=12)

# Agregar los pol√≠gonos del dataframe 'df' al mapa
for _, row in mapa.iterrows():
    folium.GeoJson(row['geometry'],
                   tooltip=f"{row['cvegeo']}<br>grado de marginacion: {row['gmu']}"
                  ).add_to(m)
# Mostrar el mapa
m.show_in_browser()
```
