```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
# agrupa por latitud y longitud, y calcula la media
grouped = df.groupby('fecha').mean().reset_index()
grouped = grouped.rename(columns={'t_mean_2m_12h': 'temperatura'})
grouped = grouped.rename(columns={'relative_humidity_mean_2m_12h': 'humedad'})
grouped = grouped.rename(columns={'precip_12h': 'lluvia'})
grouped.head()
```

```python
# genera una grafica que tenga en x la columna fecha y en y la columna temperatura
fig = px.line(grouped, x='fecha', y='temperatura', title='Temperatura vs Fecha')
fig.show()
```

```python
fig = px.line(grouped, x='fecha', y='humedad', title='Humedad vs Fecha')
fig.show()
```

```python
fig = px.line(grouped, x='fecha', y='lluvia', title='Lluvia vs Fecha')
fig.show()
```
